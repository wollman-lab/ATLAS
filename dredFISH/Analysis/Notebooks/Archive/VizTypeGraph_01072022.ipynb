```python
Vcellmap = CellMap(TMG)
Vcellmap.figsize = (10,10)
Vcellmap.cmap_list = ['YlOrBr','RdPu','PuBu','YlGn']
Vcellmap.set_view()
Vcellmap.plot()

# TMG.plot_cond_entropy_at_different_resolutions(0)
```
