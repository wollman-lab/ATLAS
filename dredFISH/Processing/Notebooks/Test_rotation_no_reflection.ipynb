```python
""" Check Reference Registration """
import numpy as np
import matplotlib.pyplot as plt 
import nrrd

from dredFISH.Utils import regu
# from dredFISH.Utils import imageu
from dredFISH.Processing.Section import *

import importlib
importlib.reload(regu)
# importlib.reload(imageu)
```

```python
theta = 2*np.pi*np.hstack([np.linspace(0,  5/16, 100), 
                           np.linspace(7/16,  1, 100)
                         ])
r = 10 # np.random.randn(100)

x1 = r*np.cos(theta)
y1 = r*np.sin(theta)

x2 = -2*r+2*r*np.cos(theta)
y2 = -2*r+2*r*np.sin(theta)

# raw
x = np.hstack([x1, x2])
y = np.hstack([y1, y2])
xy = np.vstack([x,y]).T

fig, ax = plt.subplots() 
ax.scatter(x, y)
ax.set_aspect('equal')
```

```python
# rotate
regdata = regu.RegData('test', xy)
regdata.pca_rotate()
xyrot = regdata.points_rot

fig, ax = plt.subplots() 
ax.scatter(xyrot[:,0], xyrot[:,1])
ax.set_aspect('equal')
```

```python
# rotate
regdata = regu.RegData('test', xy)
regdata.pca_rotate(allow_reflection=True)
xyrot = regdata.points_rot

fig, ax = plt.subplots() 
ax.scatter(xyrot[:,0], xyrot[:,1])
ax.set_aspect('equal')
```
