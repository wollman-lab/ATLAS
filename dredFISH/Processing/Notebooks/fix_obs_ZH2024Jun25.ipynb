```python
from dredFISH.Processing.execute import *
project_path = '/scratchdata1/Images2024/Zach/MouseBrainAtlas/'
datasets = [dataset for dataset in os.listdir(project_path) if dataset.split('_')[0] in ['WTM01','WTF01','MMSM01','MMSF01']]
datasets = [dataset for dataset in datasets if len([i for i in os.listdir(os.path.join(project_path,dataset)) if 'ybe7' in i]) > 0]
datasets
```

```python
import gc
```

```python
cword_config = 'dredfish_processing_config_tree'
config = importlib.import_module(cword_config)
for dataset in datasets:
    metadata_path = os.path.join(project_path,dataset)
    image_metadata = Metadata(metadata_path)
    hybe = [i for i in image_metadata.acqnames if config.parameters['nucstain_acq']+'_' in i.lower()]
    posnames = np.unique(image_metadata.image_table[np.isin(image_metadata.image_table.acq,hybe)].Position)
    sections = np.unique([i.split('-Pos')[0] for i in posnames if '-Pos' in i])
    for section in tqdm(sections,desc=dataset):
        # print(dataset, section)
        gc.collect()
        self = Section_Class(metadata_path,section,cword_config,verbose=False)
        self.setup_output()
        self.load_metadata()
        self.model_type = 'total'
        self.data = self.load(file_type='anndata',model_type=self.model_type)
        if self.data.obs['image_x'].max() > 0:
            del self
            gc.collect()
            continue
        self.parameters['overwrite'] = True
        self.stitcher('hybe24','FarRed')
        self.segment()
        self.parameters['vector_overwrite'] = True
        self.pull_vectors()
        del self
        gc.collect()
```
