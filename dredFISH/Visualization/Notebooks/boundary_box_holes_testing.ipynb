```python
import matplotlib.patches as patches

def add_polygon_patch(coords, ax, fc='blue'):
    patch = patches.Polygon(np.array(coords.xy).T, fc=fc)
    ax.add_patch(patch)
```

```python
XY = TMG.Layers[0].XY 
d, bb = bounding_box(XY, fill_holes=True)
fig, ax = plt.subplots(1, 1)

add_polygon_patch(bb.exterior, ax)
for interior in bb.interiors:
    add_polygon_patch(interior, ax, 'white')
        
ax.axis('equal')
plt.show()
```
