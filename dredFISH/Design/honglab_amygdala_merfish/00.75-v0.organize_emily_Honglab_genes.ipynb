```python
import pandas as pd 
```

```python
f = "/bigstore/GeneralStorage/fangming/projects/amygdala/Honglab_Amygdala_genes_Nov10_2022.csv"
df = pd.read_csv(f)
df
```

```python
df2 = df.copy()
df2['priority_binary'] = (df2['priority'] == "Yes").astype(int)
df2['Gene_name_std'] = df2['Gene name'].apply(lambda x: x[0].upper()+x[1:].lower())
```

```python
genes_rename = {
    'Fam46a': "Tent5a", 
    'Is11': "Isl1",
    'Nov': "Ccn3",
    'Papp2': "Pappa2",
    'Park2': "Prkn",
    'Cdh33': "Cdh23", 
}

def rename_genes(g):
    """
    """
    if g in genes_rename.keys():
        return genes_rename[g]
    else:
        return g
    
df2['Gene_name_std'] = df2['Gene_name_std'].apply(rename_genes)
```

```python
df2 = df2.groupby('Gene_name_std').sum(numeric_only=False)
df2['priority_binary'] = (df2['priority_binary'] > 0).astype(int)
df2 = df2.reset_index()
df2 = df2.sort_values(['priority_binary', 'Gene_name_std'], ascending=[False, True], ignore_index=True)
df2
```

```python
f2 = "/bigstore/GeneralStorage/fangming/projects/amygdala/Honglab_Amygdala_genes_organized_Nov15_2022.csv"
df2.to_csv(f2, header=True, index=True)
```
