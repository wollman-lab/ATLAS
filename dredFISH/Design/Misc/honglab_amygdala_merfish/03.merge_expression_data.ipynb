# merge

```python
tmp = df2.T.copy()
tmp['clst'] = df1.iloc[:,0].values
tmp
```

```python
merged = tmp.groupby('clst').sum()
merged
```

```python
cov_merged = merged.sum(axis=1)
```

```python
n_merged = merged.divide(cov_merged, axis=0)
```

```python
z_merged = (n_merged - n_merged.mean(axis=0))/n_merged.std(axis=0)
```

```python
sns.clustermap(z_merged, cmap='coolwarm')
```
