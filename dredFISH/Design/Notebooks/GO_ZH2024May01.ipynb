# Expected Signal

```python
import importlib
config = importlib.import_module('dredfish_processing_config_tree')
bitmap = config.bitmap
```

```python
from dredFISH.Processing.Section import *
adata = anndata.read_h5ad('/orangedata/ExternalData/Allen_WMB_2024Mar06/projected_Tree_reorder_2024Mar06/10X_combined.h5ad')
adata
```

```python
labels = adata.obs['subclass'].copy()
sample = []
for label in np.unique(labels):
    sample.extend(list(np.random.choice(np.where(labels==label)[0],1000)))
adata = adata[sample,:]
adata
```

```python
adata.X  = np.log10(np.clip(adata.X,1,None))
adata.X = adata.X * (np.mean(np.sum(adata.X,axis=1,keepdims=True))/np.sum(adata.X,axis=1,keepdims=True))
```

```python
print(adata.obs['class'].unique().shape,adata.obs['supertype'].unique().shape,adata.obs['subclass'].unique().shape)
```

```python
X = adata.X.copy()
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Group the data by labels and calculate the mean for each group
grouped_data = pd.DataFrame(np.array(adata.X),columns=[h for r,h,c in bitmap if 'RS' in r])
# grouped_data['label'] = np.array(labels)
avg_grouped_data = grouped_data.groupby(np.array(adata.obs['subclass'])).mean()
# Calculate the z-score of avg_grouped_data across columns
from scipy.stats import zscore
avg_grouped_data = avg_grouped_data.apply(zscore, axis=0)
# Clipping the avg_grouped_data values between -2 and 2
avg_grouped_data = avg_grouped_data.clip(lower=-3, upper=3)
for class_label in adata.obs['class'].unique():
    converter = dict(zip(adata.obs['subclass'],adata.obs['class']))
    subclasses = [i for i in adata.obs['subclass'].unique() if converter[i]==class_label]
    print(class_label,len(subclasses))
    if len(subclasses)>1:
        sns.clustermap(avg_grouped_data.loc[subclasses,:], cmap='bwr',col_cluster=False)
        plt.title(f"10X scRNAseq Expectations {class_label}")
        plt.xlabel('Features')
        plt.ylabel('Cell Types')
        plt.show()



```
